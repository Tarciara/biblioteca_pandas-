```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('varejo.xlsx')
vendas.head()

#lembrando que o .head() me mostra as 5 primeiras linhas da base de dados varejo.xlsx
```

```python
#analisando um pouco melhor a base de dados varejo, utilizando a função info():

vendas.info()

#observe que para preço e para estado, eu tenho uma quantidade de não nulos inferiores que as demais colunas (que tem tudo preenchido = RangeIndex). Isso significa que tenho dados nulos no meu dataset
```

```python
#nunique() para saber a quantidade por categoria (aqui, idcanalvenda)

vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
#APP e Aplicativo são o mesmo canal de venda. Vamos unir os dois, usando a função replace

vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
# novamente nunique() para saber a quantidade por categoria (aqui, bandeira)

vendas.groupby('bandeira').idcompra.nunique()
```

```python
#nunique agora para ver a quantidade por nome do departamento

vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
#nunique agora para ver a quantidade por estado

vendas.groupby('estado').idcompra.nunique()
```

```python
#usando a função replace para colocar o underline nos nomes das subcategorias da coluna nome do departamento

vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ', '_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
#vamos tratar os dados nulos

vendas.info()
```

```python
vendas[vendas['estado'].isnull()]

#conseguimos agora visualizar os dados de estado que são nulos na tabela
```

```python
#vamos eliminar (dropar) todas as linhas nulas

vendas_sem_nulos = vendas.dropna()   #dropna quer dizer: delete os nulos
vendas_sem_nulos.info()

# mas veja que eliminou as linhas das outras colunas também
```

```python
# é uma premissa dessa análise que os dados nulos sejam substituidos por MT

vendas['estado'].fillna('MS', inplace = True)
vendas.info()

#agora sim, não vou ter mais dados nulos para estado. So para preço
```

```python
media_preco = vendas['Preço'].mean()
vendas['Preço'].fillna(media_preco, inplace = True)   #media_preco não está entre aspas pois é uma variável e nao uma coluna

media_preco
```

```python
vendas.info()

#agora eu tenho todas as linhas preenchidas e eu não deletei nenhuma (substituí os preços nulos pelo preço médio, usando fillna)
```

```python
# também poderia ser substituido pela mediana (ponto médio que divide os preços):

mediana_preco = vendas['Preço'].median()
mediana_preco

#metade dos dados é menor que o valor abaixo e a outra metade é maior
```

```python
#posso substituir, por exemplo, numa premissa dada, todos os valores nulos por 1000:

vendas['Preço'].fillna(1000, inplace = True)   #esse seria o código

#aqui foi só uma introdução a tratamento de dados nulos. Mais adiante, em data clean, veremos melhor como tratar dados nulos

#até aqui já verificamos, tratamos os dados, eliminamos os nulos, corrijimos colunas que não estavam coerentes. A partir de agora, vamos analisar para tomar decisões. Explorar para entender nossos dados
```

```python
vendas[['Preço', 'Preço_com_frete']].describe()

#o desvio padrão (std) é maior que a média. Quer dizer que tenho mais dados espalhados entre mais baratos e mais caros
```

```python
# uma das premissas é de que o preço não seja maior do que o preço com frete

preco_errado = vendas.query('Preço > Preço_com_frete')  #qudo houver um preço maior que o preço com frete, tenho um preço errado

vendas_correto = vendas.query ('Preço < Preço_com_frete')

vendas_correto.describe()

#podemos ver que temos bastante preço errado, pois na tabela logo acima (onde tem todos os preços certos e errados) tem cerca de 80 mil e aqui abaixo, 22 mil, que sao os preços corretos
```

```python
vendas_correto.head() #essa agora é a nossa nova base de dados (pq é premissa que não se quer preços maiores que preço com frete)
```

```python
# vamos fazer alguns filtros

vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
#primeira métrica pedida: contagem de departamentos mais vendidos (nunique)

vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# média de preços com frete por nome do departamento

vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index()
```

```python
# apenas duas casas decimais:

round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
#Quantidade de vendas por mês

vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False)
#aqui ainda está por dia
```

```python
#agrupando por mês, primeiro vou criar uma coluna nova chamada 'mês'

vendas_correto['mês'] = vendas_correto['Data'].dt.month
vendas_correto.head()
```

```python
#agora, faço um groupby por mês

vendas_correto.groupby('mês').idcompra.nunique().sort_values(ascending=False)
```

```python
#vamos explorar agora a segunda base de dados:

cliente = pd.read_excel('cliente_varejo.xlsx')
cliente.head()
```

```python
#o que ela tem de semelhante com vendas_correto? vamos plotar vendas_correto para analisar:

vendas_correto.head()

#e percebemos que temos cliente_Log para as duas, logo, podemos fazer uma união entre ambas (join)
```

```python
#vamos analisar na tabela cliente, idade e renda:

cliente[['idade', 'renda']].describe()
```

```python
#verificar se temos dados nulos

cliente.info()

#não temos dados nulos: 26425 entradas e o mesmo valor de Non-Null Count
```

```python
# transformar renda pra float

cliente = cliente.astype({'renda': "float"})
cliente.info()
```

```python
#join

vendas_cliente = vendas_correto.merge(cliente, how = "left", on = "cliente_Log")
vendas_cliente.head()
```

```python
#próximas métricas:
#média de renda para cada tipo de canal de venda
#média de idade de clientes por bandeira

#para isso, precisamos fazer alguns agrupamentos:

agg_idcanal_renda = vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').reset_index()
agg_idcanal_renda
```

```python
#duas casas decimais

agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

podemos verificar que a maior média de renda dos clientes por canal de venda é na categoria aplicativo. Com médias de renda muito sememlhantes entre as categorias.

```python
#média de idade dos clientes por bandeira

agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending = False).reset_index(),2)
agg_idade_bandeira
```

também muito parecidas ambas as médias de idade de clientes por bandeira

Agora teremos a cereja do bolo: a apresentação desses dados por meio das bibliotecas de data visualization

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color = 'green')

plt.ylabel('Média de Idade')

plt.title('Média de Idade por Bandeira', loc='center', fontsize = 18, fontweight = 'bold', color = 'black');
```

```python
fig = plt.figure(figsize=(7,4))

plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color = 'green')

plt.ylabel('Renda média')

plt.title('Renda média por canal de venda', loc='center', fontsize = 18, fontweight = 'bold', color = 'black');

#lembrando que o ponto e vírgula no final serve para tirar os caracteres que aparecem acima do gráfico
```
